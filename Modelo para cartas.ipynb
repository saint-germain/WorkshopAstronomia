## Código para crear cartas para el II Workshop de Astronomía en los Andes

```python
from tempfile import mkstemp
from shutil import move
from os import close,remove
    
def replaceagn(file_path, pattern, subst):
    #Create temp file
    fh, abs_path = mkstemp()
    with open(abs_path,'w') as new_file:
        with open(file_path) as old_file:
            for line in old_file:
                new_file.write(line.replace(pattern, subst))
    close(fh)
    #Remove original file
    remove(file_path)
    #Move new file
    move(abs_path, file_path)
```

```python
import pandas as pd
df = pd.read_csv('charlas.csv', index_col=0)
b='formal_letter_2.tex'
tot=len(df)
```

```python
for k in range(0,tot):
    if df.WL.iloc[k]==0:
        c=df.Apellido.iloc[k][0]+'_'+df.Correo.iloc[k][0:5]+'_'+str(int(df.index[k]))
        cname='carta_%s.tex' %c
        !cp $b $cname
        replaceagn(cname,'TITULO',df.Titulo.iloc[k])
        replaceagn(cname,'PREFIJO',df.Prefijo.iloc[k])
        replaceagn(cname,'NOMBRE',df.Nombre.iloc[k])
        replaceagn(cname,'APELLIDO',df.Apellido.iloc[k])
        replaceagn(cname,'INSTITUCION',df.Instituto.iloc[k])
        replaceagn(cname,'PREGEN',df.Pregen.iloc[k])
        replaceagn(cname,'CHOP',df.CHOP.iloc[k])
        !pdflatex $cname
```
